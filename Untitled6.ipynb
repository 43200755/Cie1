```python
import pandas as pd
df=pd.read_csv('/content/mtcarp.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
auto_count = (df['am'] == 0).sum()
manual_count = (df['am'] == 1).sum()
labels = ['Automatic', 'Manual']
counts = [auto_count, manual_count]
plt.bar(labels, counts, color=['blue', 'red'])
plt.title("Frequency of Transmission Types")
plt.xlabel("Transmission Type")
plt.ylabel("Number of Cars")
plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['wt'], df['mpg'], color='yellow')
plt.title("Car Weight vs MPG")
plt.xlabel("Weight (1000 lbs)")
plt.ylabel("Miles per Gallon")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
auto = (df['am'] == 0).sum()
manual = (df['am'] == 1).sum()
plt.bar(['Automatic', 'Manual'], [auto, manual], color=['orange', 'blue'])
plt.title("Frequency of Transmission Types")
plt.ylabel("Number of Cars")
plt.grid(axis='y')
plt.show()
```

```python
plt.boxplot(df['mpg'])
plt.title("Box Plot of MPG")
plt.ylabel("Miles per Gallon")
plt.grid(True)
plt.show()
```
